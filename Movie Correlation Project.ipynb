```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)    #Adjust the configuration of the plots we will create


#Read in the Data

df = pd.read_csv(r'C:\Users\ST\Desktop\Correlation in Python\movies.csv')
df
```

```python
#Let's look at the Data

df.head()
```

```python
#Let's see if there is any missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.isnull().sum()   #this will give total number of nulls for every column/variable
```

```python
# Data types for our columns

df.dtypes
```

```python
df['budget']
```

```python
#Change Data type of Columns

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
df['votes'] = df['votes'].fillna(0).astype('int64')
```

```python
df
```

```python
#create correct year column

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)     #to display all the values

df = df.sort_values(by=['gross'], inplace=False, ascending=False)   
```

```python
#Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Films')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False) 
df.head()
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
#Let's start looking at correlation. Different types of correlation: pearson, kendall, spearman. By default pearson is used.

df.corr()
```

```python
#High correlation between budget and gross
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)    #annot = Annotation

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')


plt.show()
```

```python
#Brighter color = High correlation
```

```python
#Looks at Company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)    #annot = Annotation

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')


plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
